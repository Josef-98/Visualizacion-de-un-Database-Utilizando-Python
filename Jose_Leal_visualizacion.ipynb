```python
def filter_properties_from_csv(operation = 'sale', country = 'es', language = 'es', max_items = 50, property_type = 'homes', order = 'priceDown',
                               center = '40.4167,-3.70325', distance = 60000, sort = 'desc', bankOffer = 'false', maxprice = 1000000, file_path = r'C:\\Users\\JoseLeal\\Desktop\\Datos de idealista.csv'):
    '''
    Esta función filtra las propiedades del archivo CSV según los parámetros especificados.
    
    operation: Tipo de operación (venta, alquiler)
    country: País de búsqueda (es, it, pt)
    language: Idioma de búsqueda (es, it, pt, en, ca)
    max_items: Máximo de elementos a devolver
    property_type: Tipo de propiedad (viviendas, oficinas, locales, garajes, habitaciones)
    order: Orden de las listas ('priceDown', 'priceUp')
    center: Coordenadas del centro de búsqueda -> Por defecto= CENTRO DE MADRID.
    distance: Distancia máxima desde el centro.
    sort: Cómo ordenar los elementos encontrados.
    bankOffer: Si el propietario es un banco ('true', 'false')
    maxprice: Precio máximo de las propiedades.
    '''
    
    # Leer los datos desde el archivo CSV
    data = pd.read_csv(file_path)
    
    # Filtrar los datos según los parámetros de entrada
    filtered_data = data[
        (data['operation'] == operation) &               # Filtrar por operación (venta o alquiler)
        (data['propertyType'] == property_type) &         # Filtrar por tipo de propiedad
        (data['price'] <= maxprice) &                     # Filtrar por precio máximo
        (data['country'] == country) &                    # Filtrar por país (si está en el CSV)
        (data['language'] == language)                    # Filtrar por idioma (si está en el CSV)
    ]
    
    # Limitar la cantidad de resultados según max_items
    filtered_data = filtered_data.head(max_items)
    
    # Ordenar los datos según el parámetro 'order'
    if order == 'priceDown':
        filtered_data = filtered_data.sort_values(by='price', ascending=False)
    elif order == 'priceUp':
        filtered_data = filtered_data.sort_values(by='price', ascending=True)
    
    # Mostrar los primeros registros filtrados
    return filtered_data
```

```python
# Si el valor es NaN, asumamos que es 0.
df1.parkingSpace.fillna('0',inplace=True)
parking = df1.parkingSpace.str.split(",", expand = True)
    
# Renombrando columnas
parking = parking.rename(columns={0: 'hasParkingSpace', 1: 'IncludedInPrice'})
    
# si no es 0, es 1
parking['hasParkingSpace'] = parking['hasParkingSpace'].replace(["{'hasParkingSpace': True"], '1')
    
# A int
parking['hasParkingSpace'] = parking['hasParkingSpace'].astype(int)
    
# Si lo incluye es 1 si no 0
parking['IncludedInPrice'].fillna('0',inplace=True) # los Nones no lo está.
parking['IncludedInPrice'] = parking['IncludedInPrice'].replace([" 'isParkingSpaceIncludedInPrice': True}"], '1')
parking['IncludedInPrice'] = parking['IncludedInPrice'].replace([" 'isParkingSpaceIncludedInPrice': False"], '0')
    
# A int
parking['IncludedInPrice'] = parking['IncludedInPrice'].astype(int)
    
# Eliminar la ultima columna
parking.drop([2], axis=1, inplace=True)
```

Finalmente, reemplazamos los valores de la columna en el DataFrame original.

```python
df['parkingSpace'] = parking['hasParkingSpace']
```

```python
from folium.plugins import FastMarkerCluster

# Obtener las listas de latitudes y longitudes
lats2018 = df['latitude'].tolist()
lons2018 = df['longitude'].tolist()

# Crear una lista de ubicaciones (tuplas de latitudes y longitudes)
locations = list(zip(lats2018, lons2018))

# Crear el mapa de folium centrado en Madrid
map1 = folium.Map(location=[40.41831, -3.70275], zoom_start=11.5)

# Añadir FastMarkerCluster con las ubicaciones
FastMarkerCluster(locations).add_to(map1)

# Mostrar el mapa
map1
```
